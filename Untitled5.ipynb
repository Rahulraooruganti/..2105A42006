```python
     import numpy
X=3
Y=2
eta=0.01
eps=0.000001
del_x=3
del_y=2
max_iters=10000
iters=0
def deriv(x,y):
  x_deriv=6*(x)
  y_deriv=-5*(numpy.exp(-y))
  return  x_deriv,y_deriv
while max(abs(del_x),abs(del_y)) > eps and iters < max_iters:
  prev_x=X
  prev_y=Y
  del_x,del_y=deriv(prev_x,prev_y) 
  del_x=-eta*del_x
  del_y=-eta*del_y
  X=X+del_x
  Y=Y+del_y
  iters=iters+1
  print("iteration",iters,"\n x values is",X,"\n y value is",Y)
print("local minimum occurs at",X,Y)
```
